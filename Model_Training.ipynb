```python
import pandas as pd

df = pd.read_csv(
    "https://github.com/YBIFoundation/ProjectDataSet/raw/main/SentimentAnalysisTechNovaMobile.csv"
)

df.head()
```

```python
x = df['Comment']
y = df['Sentiment']

print(x.head())
print()
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)

print(len(x_train))
print(len(x_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

x_train_vec = tfidf.fit_transform(x_train)
x_test_vec = tfidf.transform(x_test)

print(x_train_vec.shape)
print(x_test_vec.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(x_train_vec, y_train)

print("Model Trained Successfully")
```

```python
y_pred = model.predict(x_test_vec)

print(y_pred[:10])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import pickle

pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(tfidf, open("vectorizer.pkl", "wb"))

print("Model Saved Successfully")
```

```python
review = ["Camera quality is amazing"]

review_vec = tfidf.transform(review)

prediction = model.predict(review_vec)

print("Sentiment:", prediction[0])
```

```python
confidence = model.predict_proba(review_vec)

print("Confidence:", round(max(confidence[0])*100,2), "%")
```

```python
reviews = [
    "Battery backup is terrible",
    "Best phone in this price range",
    "Display size is 6.7 inches"
]

reviews_vec = tfidf.transform(reviews)

predictions = model.predict(reviews_vec)

for review, pred in zip(reviews, predictions):
    print(review, "=>", pred)
```
